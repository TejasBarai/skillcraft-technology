```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df =pd.read_csv("train_black12.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python

# Plotting the histogram
plt.hist(df['Age'], bins=20)

# Adding a title and labels
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Purches')

# Display the plot
plt.show()
```

```python
df['Age'].value_counts()
```

```python
df['Gender'].value_counts()
```

```python
# Creating a bar plot using seaborn
sns.barplot(x='Gender', y='Purchase', data=df, palette='viridis')

# Adding titles and labels
plt.title('Bar Plot Gender')
plt.xlabel('Gender')
plt.ylabel('Purchase')

# Display the plot
plt.show()
```

```python
df['Gender']=df['Gender'].replace({'F':0 , 'M':1})
df['Gender'].value_counts
```

```python
df['Gender'].value_counts().plot(kind='pie', figsize=(10, 6),
                              ylabel='',
                              labels=['1', '0'],  # Adjust if needed
                              shadow=True, explode=[0.1,0.1] ,  # Adjust explode list to match the number of labels
                              legend=True, autopct='%1.2f%%')

plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.title('Age')
plt.show()
```
